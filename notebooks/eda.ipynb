Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Load Dataset

```python
df = pd.read_csv("../data/ai_job_replacement_2020_2026_v2.csv")

df.head()
```

Dataset Shape

```python
df.shape
```

Dataset Info

```python
df.info()
```

Statistical Summary

```python
df.describe()
```

Check Missing Values

```python
df.isnull().sum()
```

Visualization:

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

Target Variable Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["ai_replacement_score"], kde=True)

plt.title("Distribution of AI Job Replacement Risk")
plt.xlabel("AI Replacement Score")

plt.show()
```

Automation Risk vs AI Risk

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df["automation_risk_percent"],
    y=df["ai_replacement_score"]
)

plt.title("Automation Risk vs AI Replacement Score")

plt.show()
```

Skill Gap vs AI Risk

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df["skill_gap_index"],
    y=df["ai_replacement_score"]
)

plt.title("Skill Gap vs AI Risk")

plt.show()
```

Salary Change vs AI Risk

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="salary_change_percent",
    y="ai_replacement_score",
    data=df
)

plt.title("Salary Change vs AI Job Risk")

plt.show()
```

Industry vs AI Risk

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    x="industry",
    y="ai_replacement_score",
    data=df
)

plt.xticks(rotation=45)

plt.title("Industry vs AI Job Risk")

plt.show()
```

Country vs AI Risk

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    x="country",
    y="ai_replacement_score",
    data=df
)

plt.xticks(rotation=45)

plt.title("Country vs AI Job Risk")

plt.show()
```

Correlation Matrix

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

Top Features Influencing AI Risk

```python
corr["ai_replacement_score"].sort_values(ascending=False)
```
